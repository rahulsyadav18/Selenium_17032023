```python
url=[]
url1=driver.find_elements(By.XPATH,'//h1[@class="h3 lh-condensed"]/a')
for i in url1:
    url.append(i.get_attribute('href'))
```

```python
url
```

```python
lang=[]
try:
    lang1=driver.find_elements(By.XPATH,'/html/body/div[1]/div[4]/main/div[3]/div/div[2]/article/div/span/span')
    for i in lang1:
        lang.append(i.text)
except NoSuchElementException:
    lang.append('-')
            
```

```python
len(lang)
```

```python
lang0=lang[1::2]
```

```python
lang0
```

```python
lang0.insert(10,'-')
```

```python
len(lang0)
```

```python
title_re=[]
description=[]
contributors=[]
language=[]
for url in url:
    driver.get(url)
    time.sleep(2)
    try:
        title1=driver.find_elements(By.XPATH,'//*[@id="repository-container-header"]/div[1]/div/div/strong/a')
        for i in title1:
            title_re.append(i.text)
    except NoSuchElementException:
        title_re.append('-')
            
    
    try:
        title1=driver.find_elements(By.XPATH,'//*[@id="readme"]/div[2]/article/p[1]')
        for i in title1:
            description.append(i.text)
    except NoSuchElementException:
        description.append('-')
        
    try:
        contri1=driver.find_elements(By.XPATH,'//*[@id="repo-network-counter"]')
        for i in contri1:
            contributors.append(i.text)
    except NoSuchElementException:
        contributors.append('-')
    
    try:
        lang1=driver.find_elements(By.XPATH,'//span[@class="color-fg-default text-bold mr-1"]')
        for i in lang1:
            language.append(i.text)
    except NoSuchElementException:
        language.append('-')
            
```

```python
len(description),len(title_re),len(contributors),len(language)
```

```python
df4=pd.DataFrame({' Repository title': title_re,'Repository description':description,' Contributors count':contributors,'Language used':lang0})
df4
```
